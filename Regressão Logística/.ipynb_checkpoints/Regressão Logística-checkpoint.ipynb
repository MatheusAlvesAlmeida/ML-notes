# Regressão Logística com Python

A regressão logística é uma técnica estatística que tem como objetivo produzir, a partir de um conjunto de observações, um modelo que permita a predição de valores tomados por uma variável categórica, frequentemente binária, a partir de uma série de variáveis explicativas contínuas e/ou binárias. Saiba mais em [Wikipédia](https://pt.wikipedia.org/wiki/Regress%C3%A3o_log%C3%ADstica).

Nesse arquivo eu irei fazer um modelo de predição para uma competição introdutória do site [Kaggle](https://www.kaggle.com/c/titanic) usando os conceitos básicos de regressão logística. O objetivo é classificar os passageiros como vivos ou mortos de acordo com os dados apresentados no dataset.

## Preparação dos dados

Após fazer algumas leituras dos gráficos que relacionam as colunas do dataset, vou preparar os dados para que o modelo seja o mais preciso possível. 

Foi visto que há alguns dados faltando na coluna 'Age', mas deletar essa coluna devido a quantidade pequena de dados faltando não é uma boa ideia. Há várias formas de preencher esses dados. Aqui eu os preenchi com a média da idade da classe em que o passageiro se encontra. Por exemplo, se está faltando a informação de um passageiro da 1a classe, sua idade será preenchida com a média das idades da 1a classe e assim para as demais. 

Já a coluna 'Cabin' possui uma quantidade significativa de dados desconhecidos, a melhor forma de tratar isso é deletando a coluna do dataset. 

Também há uma linha na coluna de embarque que não possui informação. Nesse caso, apenas irei deletar essa linha.

```python
plt.figure(figsize=(12, 7))
sns.boxplot(x = 'Pclass', y = 'Age', data = df)
```

```python
def preencherIdades(cols):
    idade = cols[0];
    pClass = cols[1];
    
    if pd.isnull(idade):
        if(pClass == 1):
            return 37;
        elif (pClass == 2):
            return 29;
        else:
            return 24;
    else:
        return idade;
```

```python
df['Age'] = df[['Age','Pclass']].apply(preencherIdades, axis = 1)
```

```python
plt.figure(figsize = (12,6))
sns.heatmap(df.isnull())
```
